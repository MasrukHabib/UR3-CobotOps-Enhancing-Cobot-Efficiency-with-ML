```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller

# # Sample data
# # Replace this with your actual data loading process
# data = {
#     'Timestamp': pd.date_range(start='2022-10-26 08:17:21.847000', periods=100, freq='D'),
#     'Tool_current': [i + (i % 10) for i in range(100)]  # Example data
# }

# # Create a DataFrame
# df = pd.DataFrame(data)

# # Set the Timestamp column as the index
# df.set_index('Timestamp', inplace=True)

# Perform the Augmented Dickey-Fuller test on the 'Tool_current' column
result = adfuller(dfa['Tool_current'])

# Extract the test statistic, p-value, and critical values
test_statistic = result[0]
p_value = result[1]
critical_values = result[4]

# Print the results
print(f'ADF Statistic: {test_statistic}')
print(f'p-value: {p_value}')
print('Critical Values:')
for key, value in critical_values.items():
    print(f'   {key}: {value}')

# Interpretation
if p_value < 0.05:
    print("The time series is stationary.")
else:
    print("The time series is not stationary.")
```

```python
# prompt: find best parameter of p,d, q in the arima model also  pridect forcasting and show properlly and more clearlly visualised in here forecasting not clear try to make more cliar and also give only few record of forcasting

from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import mean_squared_error
import pmdarima as pm

# Assuming 'dfa' is your DataFrame with 'Tool_current' as a column and datetime index

# 1. Determine the order of differencing (d)
# Plot ACF and PACF to identify the order of differencing needed for stationarity
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
plot_acf(dfa['Tool_current'], ax=axes[0])
plot_pacf(dfa['Tool_current'], ax=axes[1])
plt.show()

# If the ACF plot shows slow decay, it might suggest non-stationarity.
# Perform the Augmented Dickey-Fuller test to confirm.
result = adfuller(dfa['Tool_current'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
# If p-value is greater than 0.05, consider differencing.

# 2. Find optimal (p, q) values
# Use auto_arima to find the best (p, d, q) values
model = pm.auto_arima(dfa['Tool_current'], start_p=1, start_q=1,
                      test='adf',       # use adftest to find optimal 'd'
                      max_p=5, max_q=5, # maximum p and q
                      m=1,              # frequency of series (assuming daily data)
                      d=None,           # let model determine 'd'
                      seasonal=False,   # No Seasonality
                      start_P=0,
                      D=0,
                      trace=True,
                      error_action='ignore',
                      suppress_warnings=True,
                      stepwise=True)

print(model.summary())

# Extract the best (p, d, q) values
best_p = model.order[0]
best_d = model.order[1]
best_q = model.order[2]

# 3. Fit the ARIMA model
model = ARIMA(dfa['Tool_current'], order=(best_p, best_d, best_q))
model_fit = model.fit()

# 4. Forecasting
# Forecast a few steps ahead (e.g., 10 steps)
forecast_steps = 10
forecast_result = model_fit.get_forecast(steps=forecast_steps)
fc = forecast_result.predicted_mean
conf_int = forecast_result.conf_int()

# 5. Visualization
plt.figure(figsize=(12, 5))
plt.plot(dfa['Tool_current'], label='Actual')
plt.plot(fc, label='Forecast')
plt.fill_between(conf_int.index, conf_int.iloc[:, 0], conf_int.iloc[:, 1], color='k', alpha=.15)
plt.title('Tool Current Forecast')
plt.legend()
plt.show()

# Print the forecasted values
print(fc)
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Load your data
# Assuming 'data.csv' is your data file and it contains a column 'Tool_current'
# data = pd.read_csv('data.csv', parse_dates=True, index_col='Date')
tool_current = df['Tool_current']

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(tool_current)
plt.title('Tool Current Over Time')
plt.xlabel('Date')
plt.ylabel('Tool Current')
plt.show()

# Split data into train and test sets
train_size = int(len(tool_current) * 0.8)
train, test = tool_current[:train_size], tool_current[train_size:]

# Fit ARIMA model
# Replace p, d, q with appropriate values for your data
p = 5   # lag order
d = 1   # degree of differencing
q = 2   # size of moving average window

model = ARIMA(train, order=(p, d, q))
model_fit = model.fit()

# Summary of the model
print(model_fit.summary())

# Make predictions
start = len(train)
end = len(train) + len(test) - 1
predictions = model_fit.predict(start=start, end=end, typ='levels')

# Plot predictions vs actual values
plt.figure(figsize=(10, 6))
plt.plot(test, label='Actual')
plt.plot(predictions, label='Predicted', color='red')
plt.title('ARIMA Model Predictions')
plt.xlabel('Date')
plt.ylabel('Tool Current')
plt.legend()
plt.show()

# Calculate mean squared error
mse = mean_squared_error(test, predictions)
print(f'Mean Squared Error: {mse}')
```
